```python
import pandas as pd

from mlgear.utils import show
```

```python
data = pd.read_csv('../../../data/2019/2019-ea-survey-INTERNAL-draft10.csv')
data
```

```python
sorted(list(data.columns))
```

```python
engagement_cols = [c for c in data.columns if 'involved' in c or 'engaged' in c or 'ea_barrier' in c or 'retention' in c or c in ['gender', 'race_white', 'age', 'ea_engaged_scale']]
engagement_cols = [c for c in engagement_cols if c not in ['involved_none_of_the_above', 'involved_other', 'ea_know_drift_involved']]
engagement_cols
```

```python
data['ea_engaged_scale'].value_counts()
```

```python
(data['ea_engaged_scale'] > 3).value_counts()
```

```python
data2 = data.copy()

data2['male'] = data2['gender'].apply(lambda x: 1 if x == 'Male' else 0)
data2.drop('gender', axis=1, inplace=True)

data2['ea_engaged_scale'] = (data2['ea_engaged_scale'] > 3)

engagement_cols = [c for c in engagement_cols if c != 'gender'] + ['male']
data2 = data2[engagement_cols].fillna(0).astype(int)
data2
```

```python
data3 = data[[c for c in data.columns if 'involved' in c and 'other' not in c and 'none' not in c]].copy()
for c in data3.columns:
    data3[c] = data3[c].apply(lambda x: c.replace('involved_', '') if x == 1 else '')

data3 = data3.fillna('').apply(' '.join, axis=1).apply(lambda x: ' '.join(sorted(x.split())))
data3 = pd.DataFrame(data3)
data3.columns = ['journey']
data3 = data3.groupby('journey')['journey'].count()
data3 = pd.DataFrame(data3)
data3.columns = ['count']
data3 = data3.reset_index()
data3 = data3[data3['journey'] != ''][data3['count'] > 4].sort_values('count', ascending=False)
data3.reset_index(drop=True, inplace=True)
show(data3, max_rows=None)
```

```python
data4 = data.copy()
data4 = data4[data4['ea_engaged_scale'] > 3]
data4 = data4[[c for c in data.columns if 'involved' in c and 'other' not in c and 'none' not in c]]
for c in data4.columns:
    data4[c] = data4[c].apply(lambda x: c.replace('involved_', '') if x == 1 else '')

data4 = data4.fillna('').apply(' '.join, axis=1).apply(lambda x: ' '.join(sorted(x.split())))
data4 = pd.DataFrame(data4)
data4.columns = ['journey']
data4 = data4.groupby('journey')['journey'].count()
data4 = pd.DataFrame(data4)
data4.columns = ['count']
data4 = data4.reset_index()
data4 = data4[data4['journey'] != ''][data4['count'] > 4].sort_values('count', ascending=False)
data4.reset_index(drop=True, inplace=True)
show(data4, max_rows=None)
```
